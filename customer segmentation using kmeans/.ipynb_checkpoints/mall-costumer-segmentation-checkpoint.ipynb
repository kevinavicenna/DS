# Standardizing the Data

```python
label = LabelEncoder()
```

# Machine Learning Models

## KMeans

```python
kmeans = KMeans()
```

```python
k_fit = kmeans.fit(df)
```

```python
k_fit.n_clusters
```

## Cluster Visualization

```python
df 
```

```python
ssd = []
K = range(1,30)

for k in K:

    kmeans = KMeans(n_clusters= k).fit(df)
    ssd.append(kmeans.inertia_)
```

```python
plt.plot(K,ssd, "o-")
plt.xlabel("Farkli K Değerlerine Karşılık Uzaklık Artık Toplamları")
plt.title("Optimym Küme Sayısı İçin Elbow Yöntemi")
```

```python
kmeans = KMeans()
visu = KElbowVisualizer(kmeans,k=(2,20))
visu.fit(df)
visu.poof()
```

```python
kmeans = KMeans(n_clusters=6).fit(df)
kmeans
```

```python
Clusters = kmeans.labels_
```

```python
df["Clusters"]= Clusters
```

```python
df
```

# Cluster Hierarchy

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
```

```python
plt.figure(figsize=(15,8))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel("Observation Units")
plt.ylabel("Offsets")
dendrogram(hc_complete,
          leaf_font_size= 5);
```

```python
plt.figure(figsize=(15,8))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel("Observation Units")
plt.ylabel("Offsets")
dendrogram(hc_average,
          leaf_font_size= 5);
```

```python
plt.figure(figsize=(15,8))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel("Observation Units")
plt.ylabel("Offsets")
dendrogram(hc_complete,
          truncate_mode="lastp",
          p=10,
          show_contracted=True,
          leaf_font_size=10);
```
